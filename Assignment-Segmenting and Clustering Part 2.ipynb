```python
# import important libraries
import pandas as pd 
import requests 
from bs4 import BeautifulSoup 
```

```python
# retrieve get response from wiki page
req = requests.get("https://en.wikipedia.org/w/index.php?title=List_of_postal_codes_of_Canada:_M&oldid=1012118802") 

soup = BeautifulSoup(req.content,'lxml') 

# filter everything out except table, which is the main data that we want for the project
table = soup.find_all('table')

df = pd.read_html(str(table)) 

#create a dataframe from what we have scraped from wikipedia
neighborhood=pd.DataFrame(df[0])
```

```python
neighborhood
```

```python
#remove not assigned neighborhoods from the dataset
neighborhood_cleaned = neighborhood[neighborhood['Borough'] != 'Not assigned'].reset_index(drop=True)
neighborhood_cleaned
```

```python
#print the number of rows of the dataframe
neighborhood_cleaned.shape
```

```python
#access geospatial dataset from assignment information
df2 = pd.read_csv(r'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DS0701EN-SkillsNetwork/labs_v1/Geospatial_Coordinates.csv')
df2
```

```python
#merge two tables to obtain coordinates for orignal dataset
neighborhood_coordinates = pd.merge(neighborhood_cleaned, df2, how="inner", on=["Postal Code"])
neighborhood_coordinates
```
